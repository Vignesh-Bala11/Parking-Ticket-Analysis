```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import math
%matplotlib inline
```

```python
# Loading in Parking Tickets excel as df 
df = pd.read_excel('H:\Case-Study\Data\Parking_Tickets.xlsx')
df.head()
```

```python
print(df.dtypes)
print(df.shape)
```

```python
df = df.rename({'Neighborhood ID': "Area_ID", 'Neighborhood Area Name': 'Area_Descr'}, axis=1)
```

```python
df['Date'] = df['date_of_infraction'].apply(lambda x: datetime.datetime.strptime(str(x), '%Y%m%d'))
df.head()
```

```python
# Dropping Mising addresses
df_1 = df[pd.notnull(df['location2'])]
df_1 = df_1[pd.notnull(df_1['Area_ID'])]
print(df_1.dtypes)
print(df_1.shape)
```

```python
# Dropping irrelevent columns
df_2 = df_1.drop(['tag_number_masked','location1','location3', 'location4','province','location5',], axis=1)
#df_2['Year'] = df_2['date_of_infraction'].map(lambda x: int(str(x) [0:4]))
#df_2['month'] = df_2['date_of_infraction'].map(lambda x: int(str(x) [4:6]))
#df_2['day'] = df_2['date_of_infraction'].map(lambda x: int(str(x) [6:8]))
df_2.head()
```

```python
locations = df_2['location2'].unique()
print('%s unique locations ' % len(locations))
```

```python
# Unique infraction Ids
infraction_id = list(df_2.infraction_code.unique())
print(infraction_id)
```

```python
grouped_df = pd.DataFrame({'count': df_2.groupby(['Date','infraction_code','location2','latitude','longtitude','Area_ID','Area_Descr',]).size()}).reset_index()
#grouped_df = grouped_df.set_index('UID')
print(grouped_df.shape)
grouped_df.head()
#grouped_df.sort_values(by='UID',ascending=True)
```

```python
y = np.array(grouped_df['count'])
print(np.unique(y))
```

```python
##now calculate total amount...
##create dict of infraction_code:fine_amount
fine_amount_dict={ic:{'fine_amt':df_2[df_2['infraction_code']==ic].set_fine_amount.median(),
                      'fine_descp':df_2[df_2['infraction_code']==ic].infraction_description.iloc[0]}    
                          for ic in infraction_id}


def calc_fine_sum(df):
    return df['count']*fine_amount_dict[df['infraction_code']]['fine_amt']

def get_fine_descrp(df):
    return fine_amount_dict[df['infraction_code']]['fine_descp']

grouped_df['fine_sum']=grouped_df.apply(calc_fine_sum,axis=1)
grouped_df['fine_descrp']=grouped_df.apply(get_fine_descrp,axis=1)
```

```python
# Add comment
mean_time_df=df_2.groupby(['Date','infraction_code','location2'])['time_of_infraction'].mean().reset_index()
mean_time_df.head()
merged_df=pd.merge(grouped_df, mean_time_df, on=['Date','infraction_code','location2'], how='inner')
print(merged_df.shape)
merged_df.head()
```

```python
merged_df=merged_df.rename(index=str, columns={"time_of_infraction": "mean_time_of_infraction" \
                                                               ,"location2":"street_address"})
merged_df.head()
#print(merged_df.shape)
```

```python
merged_df = merged_df.dropna()
cols = ["Date", "infraction_code", "mean_time_of_infraction" ,"Area_ID", "Area_Descr" , "street_address", "count", "fine_sum", "fine_descrp", "longtitude", "latitude"]
merged_df = merged_df.reindex(columns=cols)
print(merged_df.shape)
merged_df.head()
```

```python
# 
merged_df['new_time'] = merged_df["mean_time_of_infraction"].apply(lambda x:100 * ((x // 100) + ((x % 100) // 60)) + ((x % 100) % 60))
merged_df['new_time'] = merged_df['new_time'].apply(np.floor)
#merged_df['new_time'] = pd.to_datetime(merged_df['new_time'], format='%H%M')
print(merged_df.shape)
merged_df.sort_values('fine_sum', ascending=False).head()
```

```python
merged_df['new_time'] = merged_df.Date.astype(str) + '-' + merged_df['new_time'].astype(int).astype(str).apply(lambda x: (4 - len(x))*'0' + x)
merged_df.head()
```

```python
merged_df['new_time'] =  pd.to_datetime(merged_df['new_time'], format='%Y-%m-%d-%H%M')
merged_df.head()
```

```python
#merged_df = merged_df['new_time'].astype(str).str[10:]
merged_df['new_time'] = merged_df.new_time.astype(str).str[10:] 
merged_df.head()
```

```python
# Reordering Columns and cleaning final dataframe
final_df = merged_df.drop(['mean_time_of_infraction',"Area_Descr"], axis=1)
col_list = cols = ["Date","new_time", "infraction_code", "Area_ID", "street_address", "count", "fine_sum", "fine_descrp", "longtitude", "latitude"]
final_df = final_df.reindex(columns=col_list)
final_df['Area_ID'] = final_df["Area_ID"].astype(int)
new_name = {"new_time":'Time','street_address':'Address'}
final_df = final_df.rename(columns=new_name)
final_df.head()
```
